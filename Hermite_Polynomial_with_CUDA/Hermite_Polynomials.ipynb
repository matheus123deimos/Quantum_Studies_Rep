## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Category10
```

## Hermite Lambda Function

```python
hermite_recursion_base = lambda x, n :  1.0 if(n==0) else 2*x
HER = lambda x, n: hermite_recursion_base(x,n) if(n==0 or n==1) else 2*x*HER(x,n-1) - 2*(n-1)*HER(x,n-2)
```

## Plot Hermite Polynomials

```python
def HER_plot_1(n,y_limite):

    x_vals = np.linspace(-np.pi,np.pi,1000)

    for i in range(n):
        y_vals = []
        for x in x_vals:
            y_vals.append(HER(x,i))
        plt.plot(x_vals,y_vals,label='n = '+str(i))
    plt.legend()
    plt.grid(True)
    plt.ylim([-y_limite,y_limite])
    plt.show()
```

```python
HER_plot_1(7,1000)
```

```python
def HER_plot_2(n,y_limite):

    x_vals = np.linspace(-np.pi,np.pi,1000)
    p = figure(title='', x_axis_label='', y_axis_label='')

    for i in range(n):
        y_vals = []
        for x in x_vals:
            y_vals.append(HER(x,i))
        p.line(x_vals,y_vals, legend_label='n = '+str(i), line_width=2,color=Category10[10][i])
    output_notebook()
    show(p)
```

```python
HER_plot_2(7,1000)
```
